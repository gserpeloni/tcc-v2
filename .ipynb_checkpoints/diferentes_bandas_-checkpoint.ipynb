### Definindo Funções de uso recorrente:

````python
def mplot_openImg(img):
    mplot.imshow(img, cmap='gray', vmin=0, vmax=255)
    mplot.show()
````

````python
def cv2_histogram(img):
   return CV2.calcHist([img], [0], None, [256], [0,256])
````

````python
def show_vector_Imgs(img):
    for i in range(0, img.shape[0]):
        mplot_openImg(img[i])
````

````python
def write_Image(nome,img):
    CV2.imwrite('../tcc/'+nome,img)
````

````python
def _sift(imagem):
    # Instanciondo Operador SIFT: 
    sift = CV2.xfeatures2d.SIFT_create()
    
    #Definindo dados de entrada: 
    gray = imagem
    kp, descr_kp = sift.detectAndCompute(gray,None)

    #Desenhando os keypoints na imagem original:
    img_sift = CV2.drawKeypoints(gray,kp,imagem,flags=CV2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    
    return kp, descr_kp, img_sift
````

````python
def _harris(img_original):
    img_harris = img_original
    corner_Hr = CV2.cornerHarris(np.float32(img_harris),2,3,0.04)
    corner_Hr = CV2.dilate(corner_Hr, None) # Verificar para que serve isso 

    # Casting GRAY to RGB
    img_harris = CV2.cvtColor(img_original, CV2.COLOR_GRAY2RGB); 

    kp_harris = []

    # Aplicaçao do treshhold na imagem original - Verificar quantos pontos foram selecionados :
    img_harris[corner_Hr > (0.001 * corner_Hr.max())] = [0,255,255]

    kp_harris.append(
        img_harris[corner_Hr > (0.001 * corner_Hr.max())]    
    )

    return kp_harris, img_harris, corner_Hr
````

### 10.1 Brute-Force Matching 
- [Brute-Force.link](https://docs.opencv.org/master/dc/dc3/tutorial_py_matcher.html)

````python
# Instanciando Brute force
bf = CV2.BFMatcher()
````

> #### 10.1.1 Correlação de Pontos - SIFT

````python
#Descritores:
matches = bf.knnMatch(descr_kp_1,descr_kp_2,k=2)

# Aplicando o teste de Proporçao: 
kp_selecionados = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        kp_selecionados.append([m])


#Realizando o  matching entre as imagens : 
img_match = CV2.drawMatchesKnn(img_sift_1,kp_1,img_sift_2,kp_2,kp_selecionados,None,
                               flags=CV2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
        
#Plotando a imagem
mplot_openImg(img_match)
write_Image("../tcc/resultados/Different_BD/matching_Sift_image.jpg",img_match)
````

> #### 10.1.2 Correlação de Pontos - Harris

````python
#Descritores:
matches = bf.knnMatch(corner_Hr,corner_Hr,k=2)

# Aplicando o teste de Proporçao: 
kp_selecionados = []

for m,n in matches:
    if m.distance < 0.75*n.distance:
        kp_selecionados.append([m])

#Realizando o  matching entre as imagens : 
img_match = CV2.drawMatchesKnn(img_harris_1,kp_1,img_harris_1,kp_1,kp_selecionados,None,flags=CV2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
        
#Plotando a imagem
mplot_openImg(img_match)
write_Image("../tcc/resultados/Different_BD/matching_Harris_image.jpg",img_match)
````
